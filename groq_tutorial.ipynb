```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0,
    groq_api_key='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'
)
scrap_template=PromptTemplate.from_template("""
### SCRAPED TEXT:
{scrap}

### TASK:
You are an information extraction assistant.  
Extract all job postings from the scraped text above.

### REQUIRED OUTPUT FORMAT:
Return a *valid JSON array*.  
Each job posting must contain exactly these keys:

- "role"
- "experience"
- "skills"
- "description"

### IMPORTANT RULES:
- Do NOT include any explanation or preamble.
- Do NOT add extra fields.
- If some fields are missing, infer if possible or leave them as an empty string.
- Output ONLY valid JSON.

### FINAL OUTPUT:
Return ONLY the JSON array.
""")
raw = scrap_template|llm
fir_temp=raw.invoke({"scrap":scrap})
print(fir_temp.content)
jk=JsonOutputParser()
json_fir_temp=jk.parse(fir_temp.content)
```

```python
import pandas as pd
df=pd.read_csv("my_portfolio.csv")
print(df)
```

```python
import chromadb
import uuid
client=chromadb.PersistentClient()
collection=client.get_or_create_collection("portfolio")
if collection.count()==0:
    for _, row in df.iterrows():
        collection.add(
            documents=[row["Techstack"]],
            metadatas=[{"links":row["Links"]}],
            ids=[str(uuid.uuid4())]
        )
```

```python

links=collection.query(
    query_texts=["python"],
    n_results=2
).get("metadatas")
print(links)
```
