```python
import chromadb
Client=chromadb.Client()
collection=Client.create_collection(name="Practice")
```

```python
collection.add(
    documents=["New Delhi", "New York"],
    ids=['id1', 'id2']
              )
```

```python
collection.get()
```

```python
collection.get(ids='id1')
```

```python
collection.query(
    query_texts=["Chola Puri"],
    n_results=2
)
```
